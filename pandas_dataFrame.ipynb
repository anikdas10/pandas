```python
movie_imdb = pd.read_csv("./movies_csv_file/IMDB_INDIA.csv")
```

```python
# movies with rating higher than 8 and votes > 10000
movie_imdb[(movie_imdb["Rating"] > 8) & (movie_imdb["Vote_Count"] > 10000)]
```

```python
# Action Movies with higher than 7.5 rating
movie_imdb[
    (movie_imdb["Genre"].fillna("").str.split(",").apply(lambda x: "Action" in x))
    & (movie_imdb["Rating"] > 7.5)
]
```
