Creating DataFrame

```python
# using lists

student_data = [
    [100, 80, 10],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2],
]

pd.DataFrame(student_data, columns=["IQ", "Marks", "Package"])
```

```python
# using dict
student_dict = {
    "name": ["Anik", "Deb", "Maruf", "Rahim"],
    "iq": [100, 90, 120, 80],
    "marks": [80, 70, 100, 50],
    "Package": [10, 7, 14, 2],
}

students = pd.DataFrame(student_dict)
students.duplicated().sum()
```
